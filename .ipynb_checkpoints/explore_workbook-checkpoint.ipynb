```python
def welch_ttest(data,col):
    
    Null = "There is no difference in mean insurance charges between smokers and non smokers"
    Alternative = "There is a statistically significant difference in mean insurance charges between smokers and non smokers"

    alpha = 0.5
    smokers = data[data[col] == 'yes']['charges']
    non_smokers = data[data[col] == 'no']['charges']
    
    # Perform Welch's t-test
    t_stat, p_value = ttest_ind(smokers, non_smokers, equal_var=False)
    
    if p_value < alpha:        
        print("Reject the null hypothesis.")
        print(f"{col.capitalize()} status significantly affects insurance charges.")
    else:
        print("Fail to reject the null hypothesis.")
        print("No statistically significant difference detected.")
    
    
    print("Mean charges (Smokers):", smokers.mean())
    print("Mean charges (Non-Smokers):", non_smokers.mean())
```

```python
welch_ttest(train,'smoker')
```

```python
alpha = 0.5

H_0 = "There is no difference in mean insurance charges between smokers and non smokers"
H_a = "There is a statistically significant difference in mean insurance charges between smokers and non smokers"
```

```python
smokers = train[train['smoker'] == 'yes']['charges']
non_smokers = train[train['smoker'] == 'no']['charges']

# Perform Welch's t-test
t_stat, p_value = ttest_ind(smokers, non_smokers, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("Smoking status significantly affects insurance charges.")
else:
    print("Fail to reject the null hypothesis.")
    print("No statistically significant difference detected.")
```

```python
print("Mean charges (Smokers):", smokers.mean())
print("Mean charges (Non-Smokers):", non_smokers.mean())
```

#### Takeaways:

A Welch two-sample t-test was conducted to compare insurance charges between smokers and non-smokers. The null hypothesis stated that there is no difference in mean charges between the two groups, but after conducting our test we found that smoking does indeed significantly affect insurance charges.

```python
def pearsonr_test (data, col):
    alpha = 0.05

    Null_Hypothesis= f'There is no linear correlation between {col.capitalize()} and insurance charges' 
    Alternative_Hypothesis = f'{col.capitalize()} is positively correlated with insurance charges.'
    
    r, p_value = pearsonr(x = data[col], y = data['charges'])

    if p_value < alpha:
        print("Reject the null hypothesis.")
        print(Alternative_Hypothesis)
    else:
        print("Fail to reject the null hypothesis.")
        print(Null_Hypothesis)
    
    print("Pearson r:", r)
    print("p-value:", p_value)

```

### 4. Does the number of children significantly impact insurance charges?

```python
sns.barplot(x= train['children'], y = train['charges'])
plt.show()
```

```python
alpha = 0.05

H_0 = 'The mean insurance charges are equal across all child-count groups.'
H_a = 'At least one child-count group has a different mean insurance charge.'
```

```python
def anova_test (data, col):
    alpha = 0.05

    Null_Hypothesis= f'The mean insurance charges are equal across all {col.capitalize()}' 
    Alternative_Hypothesis = f'At least one {col.capitalize()} group has a different mean insurance charge.'

    groups= []
    
    for group in data[col].unique():
        data_point = data[data[col] == group]['charges']    
        
        groups.append(data_point)

    
    f_stat, p_value = f_oneway(*groups)

    if p_value < alpha:
        print("Reject the null hypothesis.")
        print(Alternative_Hypothesis)
    else:
        print("Fail to reject the null hypothesis.")
        print(Null_Hypothesis)
    
    print("F_stat:", f_stat)
    print("p-value:", p_value)

```

```python
group_0 = train[train['children'] == 0]['charges']
group_1 = train[train['children'] == 1]['charges']
group_2 = train[train['children'] == 2]['charges']
group_3 = train[train['children'] == 3]['charges']
group_4 = train[train['children'] == 4]['charges']
group_5 = train[train['children'] == 5]['charges']

# Run ANOVA
f_stat, p_value = f_oneway(group_0, group_1, group_2, group_3, group_4, group_5)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("The number of children significantly influences insurance charges.")
else:
    print("Fail to reject the null hypothesis.")
    print("No statistically significant difference in charges across child groups.")
```

```python
train.groupby('children')['charges'].mean()
```

#### Takeaway

- A one-way ANOVA was conducted to determine whether mean insurance charges differ across child-count groups. The null hypothesis stated that all group means are equal. We found that at least one child count group has a different mean. We see that child group 3 has the highest mean of insurance charges.

### 5. Are there statistically significant differences in charges across regions?

```python
sns.barplot(x= train['region'], y =train['charges'], color='darkgreen' )
plt.show()
```

```python
alpha = 0.05

H_0 = 'There is no difference in mean insurance charges between the regions'
H_a = "There is a statistically significant difference in mean insurance charges between regions"
```

```python
nw = train[train['region'] == 'northwest']['charges']
ne = train[train['region'] == 'northeast']['charges']
sw = train[train['region'] == 'southwest']['charges']
se = train[train['region'] == 'southeast']['charges']

f_stat, p_value = f_oneway(nw,ne,sw,se)

print('f_statistic:', f_stat)
print('p_value:', p_value)
```

```python
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("The region significantly influences insurance charges.")
else:
    print("Fail to reject the null hypothesis.")
    print("No statistically significant difference in charges across regions.")
```

```python
train.groupby('region')['charges'].mean()
```

#### Takeaway

- A one-way ANOVA indicated statistically significant differences in mean insurance charges across regions (p < 0.05). However, the magnitude of these differences is modest relative to stronger predictors such as smoking status.
