## Explore

### 1. Is smoking status a statistically significant predictor of insurance charges?

```python
e.barplot(train,x = 'smoker', color = 'lightgreen')
```

```python
e.welch_ttest(train, 'smoker')
```

#### Takeaways:

A Welch two-sample t-test was conducted to compare insurance charges between smokers and non-smokers. The null hypothesis stated that there is no difference in mean charges between the two groups, but after conducting our test we found that smoking does indeed significantly affect insurance charges.

### 2. Is BMI positively correlated with insurance costs?

```python
e.regplot(train,'bmi')
```

```python
e.pearsonr_test(train,'bmi')
```

#### Takeaways

- A Pearson correlation analysis was conducted to evaluate the linear relationship between BMI and insurance charges. There is a weak overall positive correlation.


r value:
- 0  no linear relationship
- 0.10.3  weak
- 0.30.5  moderate
- 0.5  strong

### 3. Do insurance charges increase linearly with age?

```python
e.regplot(train,'age')
```

```python
e.pearsonr_test(train,'age')
```

#### Takeways:

- A Pearson correlation analysis revealed a statistically significant but modest positive relationship between age and insurance charges (r = 0.28, p < 0.001), indicating that insurance costs tend to increase with age, though age alone is not a strong predictor.



### 4. Does the number of children significantly impact insurance charges?

```python
e.barplot(train,'children', color = 'lightblue')
```

```python
e.anova_test(train,'children')
```

#### Takeaway

- A one-way ANOVA was conducted to determine whether mean insurance charges differ across child-count groups. The null hypothesis stated that all group means are equal. We found that at least one child count group has a different mean. We see that child group 3 has the highest mean of insurance charges.

### 5. Are there statistically significant differences in charges across regions?

```python
e.barplot(train,'region', color = 'lightgreen')
```

```python
e.anova_test(train,'region')
```

#### Takeaway

- A one-way ANOVA indicated statistically significant differences in mean insurance charges across regions (p < 0.05). However, the magnitude of these differences is modest relative to stronger predictors such as smoking status.

## Preprocessing
- Encoded variables that would promote useability for my models
- Created dummy variables for categorical variables
- Added dummy variables to dataset
- Split data into subsets to model
- Scaled numeric values
- Used a log transformer to help models 

```python
X_train, y_train, X_validate, y_validate, X_test, y_test = m.model_split (train,validate, test)
X_train_scaled, X_validate_scaled, X_test_scaled = m.scaled_data(X_train, X_validate, X_test)
y_train_log, y_validate_log = m.log_tran(y_train, y_validate)
```

-----

## Modeling 

- Goal is to minimize RMSE 
- Ran over 4 models with the following algorithmns: Linear Regression, Polynomial Regression, Tweedie Regression, & Lasso Lars Regression
- Evaluated the models on train and validate data
- The best performing model then was evaluated on test data subset

```python
#baseline Model
baseline_df = m.baseline_rmse_log_and_dollars(
    y_train, y_validate,
    y_train_log, y_validate_log,
    log_transform="log1p"
)
```

```python
# models with demostration on which one performed best with lowest RMSE
models_df = m.run_regression_models(
    X_train_scaled, X_validate_scaled, X_test_scaled,
    y_train, y_validate, y_test,
    y_train_log, y_validate_log,
    baseline_df=baseline_df,
    poly_degree=2,
    log_transform="log1p",
    sort_by="Validate_Dollars"
)
models_df
```

#### Takeaways:

- The polynomial regression model achieved the lowest validation RMSE in dollar terms, indicating the best predictive performance among the evaluated models.
